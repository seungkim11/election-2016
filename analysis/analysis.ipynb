```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('../mungee/output.csv')
%matplotlib inline
```

```python
df.head()
```

```python
df.sort_values(by='retweet_count', ascending = 0).head()
```

```python
retweeted = df[df['retweet_count'] > 0]
```

```python
val = df.retweet_count.value_counts()
```

```python
val
```

```python
ddf = retweeted.groupby(['retweet_count']).agg([np.mean])
```

```python
result = pd.concat([val, ddf], axis = 1)
```

```python
result
```

```python
plt.plot(result)
plt.show()
```
